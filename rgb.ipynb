```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('rgb_new.csv')   # read our dataset
df.head() 
```

```python
df.corr(method = 'pearson')
```

```python
df.corr(method = 'spearman')
```

```python
df.corr(method = 'kendall')
```

```python
red_rl = stats.spearmanr(df["R"],  df["RL"])
red_rred = stats.spearmanr(df["R"],  df["R(Red)"])
green_gl = stats.spearmanr(df["G"],  df["GL"])
green_ggreen = stats.spearmanr(df["G"],  df["G(Green)"])
blue_bl = stats.spearmanr(df["B"],  df["BL"])
blue_bblue = stats.spearmanr(df["B"],  df["B(Blue)"])
print("R/Rl", red_rl.statistic)
print("G/GL", green_gl.statistic)
print("B/Bl", blue_bl.statistic)
print()
print("R/R(Red)", red_rred.statistic)
print("G/G(Green)", green_ggreen.statistic)
print("B/B(Blue)", blue_bblue.statistic)
```

```python
red_rl = stats.pearsonr(df["R"],  df["RL"])
red_rred = stats.pearsonr(df["R"],  df["R(Red)"])
green_gl = stats.pearsonr(df["G"],  df["GL"])
green_ggreen = stats.pearsonr(df["G"],  df["G(Green)"])
blue_bl = stats.pearsonr(df["B"],  df["BL"])
blue_bblue = stats.pearsonr(df["B"],  df["B(Blue)"])
print("R/Rl", red_rl.statistic)
print("G/GL", green_gl.statistic)
print("B/Bl", blue_bl.statistic)
print()
print("R/R(Red)", red_rred.statistic)
print("G/G(Green)", green_ggreen.statistic)
print("B/B(Blue)", blue_bblue.statistic)
```

```python
red_rl = stats.kendalltau(df["R"],  df["RL"])
red_rred = stats.kendalltau(df["R"],  df["R(Red)"])
green_gl = stats.kendalltau(df["G"],  df["GL"])
green_ggreen = stats.kendalltau(df["G"],  df["G(Green)"])
blue_bl = stats.kendalltau(df["B"],  df["BL"])
blue_bblue = stats.kendalltau(df["B"],  df["B(Blue)"])
print("R/Rl", red_rl.statistic)
print("G/GL", green_gl.statistic)
print("B/Bl", blue_bl.statistic)
print()
print("R/R(Red)", red_rred.statistic)
print("G/G(Green)", green_ggreen.statistic)
print("B/B(Blue)", blue_bblue.statistic)
```
